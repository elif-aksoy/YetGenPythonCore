# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urun_kayit():
    urun_adi = input("Ürün ismini girniz: ")
    urun_ucreti = float(input("Ürün ücretini giriniz: "))

    urun_bilgisi = f"Ürün adı: {urun_adi}\nÜrün ücreti: {urun_ucreti}TL\n"

    with open("urunler.txt", "a") as file:
        file.write(urun_bilgisi)
        file.write("\n")

    print("Ürün bilgileri urunler.txt dosyasına kaydedildi.")
urun_kayit()
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.


````python
def kelime_degistir(eski_kelime, yeni_kelime):
    # Read contents of the file
    with open("urunler.txt", "r") as file:
        urun_info = file.read()
    yenilenen_bilgi = urun_info.replace(eski_kelime, yeni_kelime)
    with open("products.txt", "w") as file:
        file.write(yenilenen_bilgi)
    print(f"Eski kelime olan {eski_kelime}, {yeni_kelime} ile urunler.txt dosyası içerisinde değiştirildi.")
kelime_degistir()
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def sur(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage

    def drive(self):
        print("Driving the vehicle.")

    def stop(self):
        print("Stopping the vehicle.")


class Bus(Vehicle):
    def sur(self, name, speed, mileage):
        super().sur(name, speed, mileage)

    def detaylar(self):
        print(f"Name: {self.name}, Speed: {self.speed} , Mileage: {self.mileage}")
````
