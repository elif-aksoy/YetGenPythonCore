# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
import time
def vki_hesapla(kilo,boy):
    vki=kilo/(boy**2)
    return round(vki,1)
def degerler(vki):
    if vki< 18.5:
        return "Zayıf"
    elif vki<=24.9:
        return "Normal kilolu"
    elif vki<=29.9:
        return "Fazla kilolu"
    elif vki<=39.9:
        return "Obez"
    elif vki<=40:
        return "İleri obez"
def liste_olusturma(isim, yas, kilo, boy, vki, deger):
    tarih=time.ctime(time.time())
    a=f"Kullanıcı adı {isim}\n"
    b=f"Kullanıcının yaş, kilo, boy değerleri {yas},{kilo},{boy}\n"
    c=f"Kullanıcının vücut kitle indeksi {vki}\n"
    d=f"Değer aralığı {deger}\n"
    e=f"Oluşturulma tarihi {tarih}\n"
    output=a+b+c+d+e
    return output
def main():
    girisSayisi=int(input("Kaç giriş yacaksınız?"))
    for i in range(girisSayisi):
        print(f"Kullanıcı kaydı {i+1}")
        isim=input("Ad soyad giriniz.")
        yas=int(input("Yaşınızı giriniz."))
        kilo=float(input("kg cinsinden kilo giriniz."))
        boy=float(input("Boyunuzu metre cinsinden giriniz."))
        vki=vki_hesapla(kilo,boy)
        deger=degerler(vki)
        output=liste_olusturma(isim,yas,kilo, boy, vki,deger)
        with open("saglik_dosyasi.txt", "a") as f:
            f.write(output)
            f.write("\n")
            print("Veri kaydedildi.")
    print("Veriler 'saglik_dosyasi.txt' 'na kaydedildi.")
main()
```

**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
menu = {
    "muz": 1.50,
    "çilek": 2.00,
    "süt": 1.75,
    "portakal": 2.50,
    "elma": 1.25,
}

def menu_goruntule():
    print("-MENÜ-")
    for i in list(menu.keys()):
        print(f"{i.upper()}:<10:{menu[i]} TL") 
def urun_fiyat_hesabi(urun,adet):
    assert adet>=0 , "Ürün adedi 0 ve 0'dan küçük olamaz."
    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun]*adet
def satinAlma():
    menu_goruntule()
    print("Almak istediğiiz ürnü giriniz. Çıkmak için 0 tuşuna basınız.")
    alinan_urunler=[]
    alinma_adet=[]
    son_fiyat=0
    while 1:
        eklemek_istenilen=input("Eklemek istediğiniz ürünü giriniz.")
        if eklemek_istenilen=="0":
            print("İşleniyor.")
            break
        eklenen_adet=int(input("Kaç adet ekleyeceğinizi giriniz."))
        fiyat=urun_fiyat_hesabi(eklemek_istenilen,eklenen_adet)
        if fiyat is None:
            print("Hatalı giriş yaptınız.")
        else:
            print(f"Satın alma ücreti: {fiyat}\nSatın almayı onaylamak için 'x' iptal etmek için 'y' giriniz.")
            onay=input("Satın almayı onaylıyor musunuz?")
            if onay=='x':
                print("Satın alma onaylandı.")
                son_fiyat +=fiyat
                alinan_urunler.append(eklemek_istenilen)
                alinma_adet.append(eklenen_adet)
                print(f"Toplam:{son_fiyat}")
            elif onay=='y':
                print("Satın alma iptal edildi.")
        return son_fiyat, alinan_urunler,alinma_adet
def main():
    print("Merhaba\nMenüyü görüntülemek için 1'e, smoothie yapmak için 2'ya çıkmak için ise 0'a basınız.")
    program=None
    while 1:
        program=int(input("Tercihiniz:"))
        if program<0 or program>2 :
            print("Yanlış tercih yaptınız. Tekrar deneyiniz.")
        elif program==0:
            print("Çıkış yapılıyor.")
            break
        elif program==1:
            menu_goruntule()
        elif program==2:
            satin_alma_ucreti, alinan_urunler, alinma_adet=satinAlma()
            with open("kayit.txt","a") as f:
                f.write(str(time.ctime(time.time())))
                f.write("\n")
                f.write(','.join(map(str, alinan_urunler)))
                f.write("\n")
                f.write(','.join(map(str, alinma_adet)))
                f.write("\n")
                f.write(f"Toplam: {satin_alma_ucreti}")
                f.write("\n")
            cikis=input("Başka bir işlem yapmak istiyor musunuz?")
            if cikis=="evet":
                pass
            else:
                print("Teşekkürler")
                break
main()
```
